### EDA

```python
bins = [18, 30, 40, 50, 60, 70, 80, 110]
labels = ['18-30','31-40','41-50','51-60','61-70','71-80','81+']

df['age_group'] = pd.cut(df['current_age'],
                         bins=bins,
                         labels=labels)

avg_income = df.groupby('age_group')['yearly_income'].mean()

plt.figure(figsize=(8,5))
plt.bar(avg_income.index.astype(str), avg_income.values)

plt.title('Average Yearly Income by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Yearly Income ($)')

plt.show()
```

Customers in the 51-60 segment have the highest average yearly income ,while the 71-80 segment have the lowest average yearly income 

```python
import matplotlib.pyplot as plt

avg_income = df.groupby('gender')['yearly_income'].mean()

plt.figure(figsize=(6,5))
plt.bar(avg_income.index, avg_income.values)

plt.title('Average Yearly Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Yearly Income ($)')

plt.show()
```

* Male customers have a higher average yearly income than female customers.
* The income difference between genders is relatively small.

```python
import matplotlib.pyplot as plt

avg_credit = df.groupby('current_age')['credit_score'].mean()

plt.figure(figsize=(12,6))
plt.plot(avg_credit.index, avg_credit.values, marker='o')

plt.title('Average Credit Score by Age')
plt.xlabel('Age')
plt.ylabel('Average Credit Score')

plt.grid(alpha=0.3)

plt.show()
```

The average credit score remains relatively stable across most age groups (18–80), generally ranging between 695 and 730. While some higher average scores appear among customers aged above 90, these fluctuations are likely due to the very small number of observations in those age groups rather than a meaningful trend. Overall, the analysis suggests that age is not a strong predictor of credit score in this dataset.

```python
import matplotlib.pyplot as plt

avg_credit = df.groupby('num_credit_cards')['credit_score'].mean()

plt.figure(figsize=(8,5))
plt.bar(avg_credit.index, avg_credit.values)

plt.title('Average Credit Score by Number of Credit Cards')
plt.xlabel('Number of Credit Cards')
plt.ylabel('Average Credit Score')

plt.show()
```

```python
df['num_credit_cards'].value_counts().sort_index()
```

Customers with 1 to 6 credit cards show a gradual increase in average credit score, suggesting that individuals managing multiple credit cards responsibly tend to have stronger credit profiles. Although customers with 7 and 8 credit cards exhibit the highest average credit scores, these groups contain relatively few customers. Similarly, the decline observed for customers with 9 credit cards should be interpreted cautiously, as this category consists of only 3 customers, making it insufficient to establish a reliable trend.

Business Recommendation

Banks could identify customers with good credit histories for premium credit card offers or higher credit limits, while still evaluating other risk factors.
